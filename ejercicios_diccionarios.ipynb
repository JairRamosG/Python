## Ejercicios de diccionarios

```python
mi_dicc = dict(name = 'Luis',
               age = 35,
               city = 'CDMX')

'''
a) profesion = ingeniero
b) modificar la edad a 40 años
c) imprimir el valor asociado a la llave city
d) borrar la llave valor prefesion = ingeniero
e) imprimir todas las llaves valor del diccionario
f) vefirivar si la llave 'city' existe en el diccionario
'''

mi_dicc['profesion'] = 'ingeniero'
mi_dicc['age'] = 40
print(f"{mi_dicc['city']}") 
del mi_dicc['profesion']


for llave, valor in mi_dicc.items():
    print(f'{llave}\t: {valor}') 

print(f"Llave 'city' en diccionario?:\t{'city' in mi_dicc.keys()}")
```

```python
## Contruir un diccionario a partir de dos listas
## Primera lista -> (llaves)
## Segunda lista -> (valores)

lista1 = [10, 20, 30, 40, 50]
lista2 = ['diez', 'veinte', 'treinta', 'cuarenta', 'cincuenta']
diccionario = {}
for key, value in zip(lista1, lista2):
    diccionario[key] = value

for llave, valor in diccionario.items():
    print(f'{llave}\t: {valor}') 
```

```python
## Eliminar todas las llaves - valor de un diccionario e imprimir el diccionario resultante
for llave, valor in diccionario.items():
    print(f"{llave}\t:{valor}")
    #del diccionario[llave]
diccionario
```

```python
## Unir dos diccionarios en uno solo
dict_1 = {10: 'diez', 20: 'veinte', 30: 'treinta'}
dict_2 = {40: 'cuarenta', 50: 'cincuenta'}

dict_1.update(dict_2)
dict_1
```

```python
### Obtener un diccionario con la frecuencia de cada distinto caracter de la cadena 'supercalifragilisticoespiralidoso'
cadena = 'supercalifragilisticoespiralidoso'
dict = {}
for c in cadena:
    if c in dict.keys():
        dict[c] +=1
    else:
        dict[c] = 1
dict
```
