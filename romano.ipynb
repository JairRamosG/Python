## Números romanos

```python
# Tabla para convertir decimal → romano
TABLA = [
    (1000, "M"), (900, "CM"), (500, "D"), (400, "CD"),
    (100, "C"),  (90, "XC"),  (50, "L"),  (40, "XL"),
    (10, "X"),   (9, "IX"),   (5, "V"),   (4, "IV"),
    (1, "I"),
]

VALORES = {"I": 1, "V": 5, "X": 10, "L": 50, "C": 100, "D": 500, "M": 1000}


def decimal_a_romano(n):
    if n <= 0 or n > 3_999_999:
        raise ValueError("El número debe estar entre 1 y 3 999 999.")

    if n <= 3999:
        return _base_romana(n)

    miles = n // 1000
    resto = n % 1000
    resultado = f"({_base_romana(miles)})"
    if resto > 0:
        resultado += _base_romana(resto)
    return resultado


def _base_romana(n):
    """Convierte un número 1-3999 a romano clásico."""
    resultado = ""
    for valor, simbolo in TABLA:
        while n >= valor:
            resultado += simbolo
            n -= valor
    return resultado

def romano_a_decimal(s):
    s = s.strip().upper()
    if s.startswith("("):
        cierre = s.index(")")
        miles_romano = s[1:cierre]
        resto_romano = s[cierre + 1:]
        total = _parsear(miles_romano) * 1000
        if resto_romano:
            total += _parsear(resto_romano)
        return total

    return _parsear(s)

def _parsear(s):
    """Convierte romano clásico a entero"""
    total = 0
    prev = 0
    for ch in reversed(s):
        val = VALORES[ch]
        total += val if val >= prev else -val
        prev = val
    return total


print("=" * 40)
print("  CONVERSOR DE NÚMEROS ROMANOS")
print("=" * 40)

while True:
    print("\n1. Decimal → Romano")
    print("2. Romano  → Decimal")
    print("3. Salir")
    opcion = input("Opción: ").strip()

    if opcion == "1":
        try:
            print('escriba el número')
            n = int(input("Número decimal: "))
            print("Resultado:", decimal_a_romano(n))
        except (ValueError, KeyError) as e:
            print("Error:", e)

    elif opcion == "2":
        try:
            print('escriba el número')
            s = input("Número romano: ")
            print("Resultado:", romano_a_decimal(s))
        except (ValueError, KeyError) as e:
            print("Error:", e)

    elif opcion == "3":
        print("Fin")
        break

    else:
        print("Opción no válida.")
```
