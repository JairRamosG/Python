### Dado el diccionario de diccionarios:

dics = {'esp':{'1':'uno','2':'dos','3':'tres'}, 

    'deutsch':{'1':'eins','2':'zwei','3':'drei'},    

    'francais':{'1':'un','2':'deux','3':'trois'},

    'english':{'1':'one','2':'two','3':'three'}}

Imprimir una tabla de traducción que tenga una forma similar a la siguiente:



```python
from tabulate import tabulate
import pandas as pd

dics = {'esp':{'1':'uno','2':'dos','3':'tres'}, 
        'deutsch':{'1':'eins','2':'zwei','3':'drei'},    
        'francais':{'1':'un','2':'deux','3':'trois'},
        'english':{'1':'one','2':'two','3':'three'}}

df = pd.DataFrame(dics)
headers = list(dics.keys())

print(tabulate(df, headers = headers, tablefmt='fancy_grid', stralign='center'))
```
