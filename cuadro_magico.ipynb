# Cuadro Magico

```python
def generar_cuadrado_magico(n: int) -> list[list[int]]:
    if n < 3 or n % 2 == 0:
        raise ValueError(f"n debe ser impar, también >= 3, y se recibió n={n}")

    cuadrado = [[0] * n for _ in range(n)]
    fila = 0
    col = n // 2
    cuadrado[fila][col] = 1

    for num in range(2, n * n + 1):
        nueva_fila = (fila - 1) % n
        nueva_col  = (col  + 1) % n

        if cuadrado[nueva_fila][nueva_col] != 0:
            nueva_fila = (fila + 1) % n
            nueva_col  = col

        cuadrado[nueva_fila][nueva_col] = num
        fila, col = nueva_fila, nueva_col

    return cuadrado
```

```python
def imprimir_cuadrado(cuadrado: list[list[int]]) -> None:
    n = len(cuadrado)
    ancho = len(str(n * n))
    separador = "+" + (("-" * (ancho + 2) + "+") * n)

    print(separador)
    for fila in cuadrado:
        print("| " + " | ".join(str(v).rjust(ancho) for v in fila) + " |")
        print(separador)
```

```python
if __name__ == "__main__":
    n = int(input("Ingresa el orden del cuadrado mágico (número impar >= 3): "))
    cuadrado = generar_cuadrado_magico(n)
    imprimir_cuadrado(cuadrado)
```
