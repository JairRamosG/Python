```python
lista = [5, 7, 3, 9, 8, 4, 2, 6, 1]
print(f'Arreglo original: {lista}')

comparaciones = 0
n = len(lista) - 1
for i in range(n):
    for j in range(n):
        comparaciones += 1
        if lista[j] > lista[j+1]:
            lista[j], lista[j+1] = lista[j+1], lista[j]
    print(f'> {lista}')

print(f'Arreglo ordenado: {lista}')
print(f'comparaciones: {n}')
```

```python
# Limitando las comparaciónes del final

lista = [5, 7, 3, 9, 8, 4, 2, 6, 1]
print(f'Arreglo original: {lista}')

comparaciones = 0
n = len(lista) - 1
for i in range(n):
    for j in range(n-i):
        comparaciones += 1
        if lista[j] > lista[j+1]:
            lista[j], lista[j+1] = lista[j+1], lista[j]
    print(f'> {lista}')

print(f'Arreglo ordenado: {lista}')
print(f'comparaciones: {n}')
```

```python
# Usando bandera

lista = [5, 7, 3, 9, 8, 4, 2, 6, 1]
print(f'Arreglo original: {lista}')

comparaciones = 0
hay_cambios = True

while hay_cambios:
    hay_cambios = False
    n = len(lista) - 1
    for i in range(n):
        for j in range(n-i):
            comparaciones += 1
            if lista[j] > lista[j+1]:
                lista[j], lista[j+1] = lista[j+1], lista[j]
                hay_cambios = True
        print(f'> {lista} *' if hay_cambios else f'> {lista}')

print(f'Arreglo ordenado: {lista}')
print(f'comparaciones: {n}')
```
