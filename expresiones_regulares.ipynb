# Tarea
#### Cómo extraer el hostname (lo que va despues del arroba)?
1. Con código Python simple
2. Con expresión regular

```python
# Python simple
x = 'From: ramosgonzalezjair@cic.ipn.mx Tur March 09:44'
for palabra in x.split():
    if '@' in palabra:
        indice = palabra.find('@')
        print(palabra[indice+1:])
```

```python
# Expresion regular
filtro = r'@(\S+)'
x = 'From: ramosgonzalezjair@cic.ipn.mx Tur March 09:44'
for palabra in x.split():
    resultado = re.findall(filtro, palabra)
    if resultado:
        print(resultado)
```
