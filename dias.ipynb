# Días de la semana

```python
DIAS = ["Sábado", "Domingo", "Lunes", "Martes", "Miércoles", "Jueves", "Viernes"]

def dia_semana(d, m, a):

    if m <= 2:
        m += 12
        a -= 1

    K = a % 100        # año dentro del siglo
    J = a // 100       # siglo

    h = (d + (13 * (m + 1)) // 5 + K + K // 4 + J // 4 - 2 * J) % 7

    return DIAS[h]

def main():
    print("=" * 40)
    print("  DÍA DE LA SEMANA (usando Zeller)")
    print("=" * 40)

    while True:
        entrada = input("\nFecha (dd/mm/aaaa) o 'salir': ").strip()
        if entrada.lower() in ("salir", "q"):
            print("FIn")
            break
        try:
            d, m, a = map(int, entrada.split("/"))
            print(f"  → {dia_semana(d, m, a)}")
        except (ValueError, IndexError):
            print("  Formato inválido. Usa dd/mm/aaaa")
```

# Pruebas

```python
casos = [
(31, 10, 2024),   # Jueves
(1,  1,  2000),   # Sábado
(14, 2,  2025),   # Viernes 
(25, 12, 2024),   # Miércoles 
(16, 9,  1810),   # Domingo
]

for d, m, a in casos:
    print(f"  {d:02d}/{m:02d}/{a} fue un {dia_semana(d, m, a)}")
```
