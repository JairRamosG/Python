```python
# Ejercicio 5 (swap)
dicc_3 = {'diez': 10, 'veinte': 20, 'treinta': 30,
          'cuarenta': 40, 'cincuenta': 50, 'sesenta': 60}

dicc_swap = {valor: llave for llave, valor in dicc_3.items()}

print(dicc_swap)
```

```python
# Ejercicio 6 (Crear diccionario a partir de otro diccionario, extraer solo algunas llaves:valor)
dicc_4 = {'a':3, 'b':2, 'c':7, 'd':5, 'e':3}
llaves_deseadas = ['b', 'c']
dicc_5 = {k: v for k, v in dicc_4.items() if k in llaves_deseadas}
print(dicc_5)
```

```python
# Forma 1
dicc_4 = {'a':3, 'b':2, 'c':7, 'd':5, 'e':3}
dicc_5_1 = {}
lista = ['b', 'c']
for k in lista:
    dicc_5_1.update({k:dicc_4[k]})
print(dicc_5_1)
```

```python
# Forma 2 (Con comprensión de diccionario)
dicc_4 = {'a':3, 'b':2, 'c':7, 'd':5, 'e':3}
dicc_5_2 = {k:dicc_4[k] for k in lista}
print(dicc_5_2)
```

```python
# Ejercicio 8 (Dado el siguiente diccionario anidado, cambiar el nombre del estudiante Juan por cualquier otro)
dicc_anidado = {
    "clase": {
        "estudiante": {
            "nombre": "Juan",
            "calificaciones": {
                "prob": 10,
                "mat": 9,
                "alg": 10
            }
        }
    }
}
```

```python
# Forma 1
dicc_anidado["clase"]["estudiante"]["nombre"] = "Ana"

print(dicc_anidado)
```

```python
# Forma 2 (Con una función)
def cambiar_nombre(diccionario, nuevo_nombre):
    diccionario["clase"]["estudiante"]["nombre"] = nuevo_nombre

cambiar_nombre(dicc_anidado, "Carlos")

print(dicc_anidado)
```
