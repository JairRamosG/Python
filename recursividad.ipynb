### Recursividad

```python
## Factorial

def factorial(n):
    if n == 0:
        # CASO BASE
        return 1
    else:
        return n * factorial(n-1)

factorial(10)    
```

```python
## Palindromo

def is_palindrome(word):
    if len(word) < 2 :
        # CASO BASE
        return True
    else:
        if word[0] != word[-1]:
            return False
        else:
            return is_palindrome(word[1:-1])

palabra = '101'
is_palindrome(palabra)

    
```
