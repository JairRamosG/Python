# Conversión de cantidades numéricas a texto

```python
UNIDADES = [
    "", "un", "dos", "tres", "cuatro", "cinco",
    "seis", "siete", "ocho", "nueve", "diez",
    "once", "doce", "trece", "catorce", "quince",
    "dieciséis", "diecisiete", "dieciocho", "diecinueve", "veinte",
    "veintiún", "veintidós", "veintitrés", "veinticuatro", "veinticinco",
    "veintiséis", "veintisiete", "veintiocho", "veintinueve"]

DECENAS = [
    "", "", "veinte", "treinta", "cuarenta", "cincuenta",
    "sesenta", "setenta", "ochenta", "noventa"]

CENTENAS = [
    "", "ciento", "doscientos", "trescientos", "cuatrocientos", "quinientos",
    "seiscientos", "setecientos", "ochocientos", "novecientos"]


def _cientos(n):
    """Convierte un número 0-999 a palabras."""
    if n == 0:
        return ""
    if n == 100:
        return "cien"
    if n == 1000:          
        return "mil"

    partes = []

    c = n // 100
    resto = n % 100

    if c > 0:
        partes.append(CENTENAS[c])

    if resto == 0:
        pass
    elif resto < 30:
        partes.append(UNIDADES[resto])
    else:
        d = resto // 10
        u = resto % 10
        if u == 0:
            partes.append(DECENAS[d])
        else:
            # "treinta y uno", "cuarenta y dos", etc.
            partes.append(f"{DECENAS[d]} y {UNIDADES[u]}")
    return " ".join(partes)



def _entero_a_palabras(n):
    """
    Convierte un entero no negativo a su representación en palabras (español MX).
    """
    if n == 0:
        return "cero"

    ESCALA = [
        ("", ""),
        ("mil", "mil"),
        ("millón", "millones"),
        ("billón", "billones"),
        ("trillón", "trillones"),
        ("cuatrillón", "cuatrillones"),
    ]

    grupos = []
    temp = n
    while temp > 0:
        grupos.append(temp % 1000)
        temp //= 1000

    if len(grupos) > len(ESCALA):
        raise ValueError("Número demasiado grande.")

    partes = []
    for i in range(len(grupos) - 1, -1, -1):
        g = grupos[i]
        if g == 0:
            continue

        texto_grupo = _cientos(g)

        if i == 0:
            # Grupo de unidades
            partes.append(texto_grupo)

        elif i == 1:
            # Miles
            if g == 1:
                partes.append("un mil") 
            else:
                partes.append(f"{texto_grupo} mil")

        else:
            # Millones, billones, etc.
            sing, plur = ESCALA[i]
            if g == 1:
                partes.append(f"un {sing}")
            else:
                partes.append(f"{texto_grupo} {plur}")
    return " ".join(partes)



def cantidad_a_letras(cantidad):

    if cantidad < 0:
        raise ValueError("La cantidad no puede ser negativa.")

    # Separar pesos y centavos
    cantidad = round(cantidad, 2)
    pesos    = int(cantidad)
    centavos = round((cantidad - pesos) * 100)

    # Pesos en letras
    if pesos == 0:
        texto_pesos = "cero"
    else:
        texto_pesos = _entero_a_palabras(pesos)

    # Capitalizar primera letra
    texto_pesos = texto_pesos[0].upper() + texto_pesos[1:]

    # "peso" o "pesos"
    moneda = "peso" if pesos == 1 else "pesos"

    return f"{texto_pesos} {moneda} {centavos:02d}/100 M.N.."



def main():
    print("=" * 55)
    print("  CONVERSOR DE CANTIDADES A LETRAS (cheques M.N.)")
    print("=" * 55)

    while True:
        entrada = input("\nCantidad (o 'salir'): $").strip()
        if entrada.lower() in ("salir", "exit", "q"):
            print("Fin")
            break
        try:
            cantidad = float(entrada.replace(",", ""))
            resultado = cantidad_a_letras(cantidad)
            print(f"  → {resultado}")
        except ValueError as e:
            print(f"  Error: {e}")
```

```python
# Ejemplos de prueba
ejemplos = [
            1_324_121.25,
            1_111_001.15,
            100.00,
            1.00,
            1_000.00,
            31.00,
            125.00,
            0.50,
            1_000_000.00,
            999_999_999.99]

print("─" * 60)
for c in ejemplos:
    print(f"  ${c:>18,.2f} es: {cantidad_a_letras(c)}")
```

```python
main()
```
