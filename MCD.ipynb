## MCD

### Por divisores comúnes

```python
def hallar_divisores(n):
    divisores = []
    for i in range(1, n+1):
        if n%i == 0:
            divisores.append(i)
    return divisores    
```

```python
def mcd_comun(list1, list2):
    print(f'Divisores 1: {list1}')
    print(f'Lista 2: {list2}')

    mcd_list = [x for x in list1 if x in list2]
    print(f'Divisores comunes: {mcd_list}')
    mcd = mcd_list[-1]
    print(f'MCD: {mcd}')
```

```python
a = 60
b = 48

d_a = hallar_divisores(a)
d_b = hallar_divisores(b)

mcd_comun(d_a, d_b)
```

### Por Euclides - Iterativo

```python
def MCD_Euclides_iter(num1, num2):
    while True:
        residuo = num1 % num2
        if residuo == 0:
            return num2
        else:
            num1, num2 = num2, residuo
```

```python
import time
```

```python
inicio = time.time()
MCD_Euclides_iter(16, 10)
fin = time.time()
print(f'tiempo: {fin - inicio} seg')
```

### Por Euclides - mejorado

```python
def MCD_Euclides_mejorado(num1, num2):
    while num1 % num2 != 0:
        num1, num2 = num2, num1 % num2
    return num2
```

```python
inicio = time.time()
MCD_Euclides_mejorado(16, 10)
fin = time.time()
print(f'tiempo: {fin - inicio} seg')
```

### Por Euclides - recursivo

```python
def MCD_Euclides_recursivo(num1, num2):
    if num2 == 0:
        return num1
    else:
        return MCD_Euclides_recursivo(num2, num1%num2)
```

```python
inicio = time.time()
MCD_Euclides_recursivo(1199092733403101, 50420370)
fin = time.time()
print(f'tiempo: {fin - inicio} seg')
```
