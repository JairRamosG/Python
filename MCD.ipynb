## MCD

### Por divisores comúnes

```python
def hallar_divisores(n):
    divisores = []
    for i in range(1, n+1):
        if n%i == 0:
            divisores.append(i)
    return divisores    
```

```python
def mcd_comun(list1, list2):
    print(f'Divisores 1: {list1}')
    print(f'Lista 2: {list2}')

    mcd_list = [x for x in list1 if x in list2]
    print(f'Divisores comunes: {mcd_list}')
    mcd = mcd_list[-1]
    print(f'MCD: {mcd}')
```

```python
a = 60
b = 48

d_a = hallar_divisores(a)
d_b = hallar_divisores(b)

mcd_comun(d_a, d_b)
```
